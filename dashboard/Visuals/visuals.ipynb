````python
# Reading in data
df = pd.read_csv('~/Downloads/ml-py-standalone/BINARY_TAGS.csv', header = [0, 1])
df_multi = pd.read_csv('~/Downloads/ml-py-standalone/BINARY_RELATIONS.csv', header = [0, 1])
df1 = pd.read_csv('/home/msid/ltb3/Downloads/ml-py-standalone/data/public/mine_vocab_1g.csv')
````

````python
#orginizing data by NEs 
items = df.I.sum().sort_values(ascending=False)[:30]
solutions = df.S.sum().sort_values(ascending=False)[:30]
problems = df.P.sum().sort_values(ascending=False)[:30]
unknowns = df.U.sum().sort_values(ascending=False)[:30]
````

## Data Checking
---------------------------
Below is are graphs of the top 30 most occuring words for each NE

````python
solGraph = solutions.plot(kind='bar', title = "Solution Tags")
solGraph.set_xlabel("Aliases")
solGraph.set_ylabel("Occurence")
````

````python
itemGraph = items.plot(kind='bar', title = "Item Tags")
itemGraph.set_xlabel("Aliases")
itemGraph.set_ylabel("Occurence")
````

````python
probGraph = problems.plot(kind='bar', title = "Problem Tags")
probGraph.set_xlabel("Aliases")
probGraph.set_ylabel("Occurence")
````

````python
unknwnGraph = unknowns.plot(kind='bar', title = "Unknown Tags")
unknwnGraph.set_xlabel("Aliases")
unknwnGraph.set_ylabel("Occurence")
````

````python
probItem = df_multi['P I'].sum().sort_values(ascending=False)[:30]
solProbGraph = probItem.plot(kind='bar', title = "Problem Item Tags")
solProbGraph.set_xlabel("Aliases")
solProbGraph.set_ylabel("Occurence")
````

````python
solItem = df_multi['S I'].sum().sort_values(ascending=False)[:30]
solProbGraph = solItem.plot(kind='bar', title = "Solution Item Tags")
solProbGraph.set_xlabel("Aliases")
solProbGraph.set_ylabel("Occurence")
````
